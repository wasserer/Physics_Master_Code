```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cm as cm
import numpy as np
import datetime

time_numeric = mdates.date2num(analyzer.timestamps)
norm = plt.Normalize(time_numeric.min(), time_numeric.max())
colors = cm.viridis(norm(time_numeric)) 

plt.figure(figsize = (9, 7), dpi = 300)
#plt.scatter(analyzer.timestamps, analyzer.PeakPosition)
plt.scatter(analyzer.timestamps, analyzer.PeakFWHM, color = colors)
plt.ylim(0, 40)#
#plt.xlim(60, 80)
#print(analyzer.temperatures[1])
```

```python
cold_rgb = (0.2, 0.4, 1.0)  # blue
warm_rgb = (1.0, 0.5, 0.0)  # orange

# Define temperature range
t_min = 50.0
t_max = 80.0

# Function to interpolate between cold and warm color based on temp
def interpolate_color(temp):
    alpha = (temp - t_min) / (t_max - t_min)
    alpha = max(0, min(1, alpha))  # Clamp between 0 and 1
    return tuple(cold_rgb[i] * (1 - alpha) + warm_rgb[i] * alpha for i in range(3))
```

```python
plt.figure(figsize = (9,7), dpi =300)
for i in range(0, len(analyzer.temperatures)):
    temp = analyzer.temperatures[i]
    color = interpolate_color(temp)
    if analyzer.labels[i] == "Light": 
        plt.plot(analyzer.wavelengths[i], analyzer.values[i] + i * 200, color = color)
plt.xlim(300,)
```
