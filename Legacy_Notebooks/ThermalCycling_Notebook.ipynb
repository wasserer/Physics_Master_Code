```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cm as cm
import numpy as np
import datetime

time_numeric = mdates.date2num(PLanalyzer.timestamps)
norm = plt.Normalize(time_numeric.min(), time_numeric.max())
colors = cm.viridis(norm(time_numeric)) 

plt.figure(figsize = (9, 7), dpi = 300)
plt.scatter(PLanalyzer.timestamps, PLanalyzer.PeakPosition, color = colors)
plt.xlabel("T")
plt.ylabel("PeakPosition")
#plt.xlim(50, 80)

plt.figure(figsize = (9, 7), dpi = 300)
plt.scatter(PLanalyzer.timestamps, PLanalyzer.PeakHeight, color = colors)
plt.plot(PLanalyzer.timestamps, PLanalyzer.temperatures)
plt.xlabel("T")
plt.ylabel("PeakHeight")
#plt.xlim(50, 80)

fig, ax1 = plt.subplots(figsize=(9, 7), dpi=300)
sc = ax1.scatter(PLanalyzer.timestamps, PLanalyzer.PeakHeight, color=colors)
ax1.set_xlabel("Timestamps")
ax1.set_ylabel("Peak Height", color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis that shares the same x-axis
ax2 = ax1.twinx()
ax2.plot(PLanalyzer.timestamps, PLanalyzer.temperatures, color='tab:red')
ax2.set_ylabel("Temperature (°C)", color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title("Peak Height and Temperature over Time")
plt.tight_layout()

plt.figure(figsize = (9, 7), dpi = 300)
plt.scatter(PLanalyzer.timestamps, PLanalyzer.PeakFWHM, color = colors)
plt.xlabel("T")
plt.ylabel("PeakFWHM")
#plt.xlim(50, 80)
```
